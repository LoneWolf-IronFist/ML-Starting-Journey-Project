# Q1

Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5,6,7,8])
y=np.array([60,62,61,58,56,57,46,63])
plt.plot(x,y)
plt.xlabel('DAYS')
plt.ylabel('SPEED')
plt.title('CAR SPEED')
plt.show()
```

Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.


```python
plt.plot(x,y,color='g',linestyle=':',marker='o',markersize='5',markerfacecolor='g')
plt.show()
```

# Q4

Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
# Sub Plots
x=np.arange(-10,10,0.01)
plt.subplot(221)
plt.plot(x,np.sin(x),'g',label='Sin Wave')
plt.title('Subplots')
plt.legend()
plt.subplot(222)
plt.plot(x,np.cos(x),'r',label='Cos-wave')
plt.legend()
plt.title('Subplots')
plt.show()

# SuperImpostion
plt.plot(x,np.sin(x),'g',label='Sin Wave')
plt.plot(x,np.cos(x),'r',label='Cos-wave')
plt.legend()
plt.title('Superimpostion')
plt.show()
```

# Q5

Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]
# Subplots
plt.subplot(211)
plt.bar(Languages,Security,label='Security',color='orange')
plt.legend()
plt.xlabel('Languages')
plt.ylabel('Parameters')
plt.title('Information')
plt.show()
plt.subplot(212)
print('\n')
plt.bar(Languages,Popularity,label='Popularity',color='blue')
plt.xlabel('Languages')
plt.ylabel('Parameters')
plt.title('Information')
plt.legend()
plt.show()
# Barh
plt.barh(Languages,Popularity,color='red')
plt.xlabel('Popularity')
plt.ylabel('Languages')
plt.title('Horizontal bar Plot')
plt.show()
```

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df.head()
```

```python
df.Sex.value_counts().plot.pie(labels=['Male','Female'])
plt.show()
```

```python
male=df[df['Sex']=='male']
plt.scatter(male['Fare'],male['Age'],color='red',marker='*',label='Male')
female=df[df['Sex']=='female']
plt.scatter(female['Fare'],female['Age'],color='yellow',marker='^',label='Female')
plt.legend()
plt.xlabel('Fare-Paid')
plt.ylabel('Age')
plt.title('Fare-Paid vs Age for Different Sex Groups')
plt.show()
```
